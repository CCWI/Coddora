### Settings

```python
iddFile = "/usr/local/EnergyPlus-22-2-0/Energy+.idd" # SET CORRECT PATH TO IDD FILE
IDF.setiddname(iddFile)

idfFile = "./input_data/base_room.idf"
zoneName = "Zone_1"
epwFile = "./input_data/DE_Munich_Theresienwiese.epw"
occupancyFile = "./input_data/simulated_occupancy_office_rooms.csv"
#occupancyFile = "./input_data/simulated_occupancy_meeting_rooms.csv"
idf_output_path = "./ep_data/room.idf"
h5_path = "./output_data/dataset_office_rooms.h5" # output path for the final dataset

generationrate_co2 = 0.0000000382  # m³/s*W
activity_level     = 108           # W/1.8m²
simulation_days    = 10
```

```python
bounds = np.array([[2, 10],             # width in m 
                   [2, 10],             # length in m
                   [2.4, 4.0],          # height in m
                   [0.000085, 0.00085], # infiltration per facade area in m³/m²s
                   [300, 500],          # outdoor CO2 in ppm
                   [0, 359],            # facade orientation in degrees
                   [1, 5]])             # max number of occupants in offices 0-4
#                  [4, 13]])            # max number of occupants in meeting rooms 4-12

winToFloor_bounds       = [0.1, 0.6] # acceptable range of ratio between window area and floor area
spacePerOccupant_bounds = [6, 22]    # m² dedicated to one occupant in offices
#spacePerOccupant_bounds = [2,  6]    # m² dedicated to one occupant in meeting rooms
```

### Latin Hypercube Sampling (LHS)

```python
lhs = LHS(xlimits=bounds, random_state=42)

x = lhs(100000)

# round values
for i in range(len(x)):
    x[i, 0] = round(x[i, 0], 3) # round to mm
    x[i, 1] = round(x[i, 1], 3) # round to mm
    x[i, 2] = round(x[i, 2], 3) # round to mm
    x[i, 3] = round(x[i, 3], 6) # round to 6 decimal places
    x[i, 4] = int(x[i, 4])      # round to integer ppm
    x[i, 5] = int(x[i, 5])      # round to integer degrees
    x[i, 6] = int(x[i, 6])      # round to integer number of occupants

# filter out samples by space-per-occupant constraint
samples = []
for s in x:
    floorArea = s[0] * s[1]
    spacePerOccupant = floorArea / s[6]
    if ((spacePerOccupant >= spacePerOccupant_bounds[0]) & 
        (spacePerOccupant <= spacePerOccupant_bounds[1])):
        samples.append(s)
                    
samples = np.array(samples)

print(samples.shape)
```

### Build Room Models

```python
def buildRoomModel(roomBuilder, params):
    roomBuilder.updateOrientation(params['orientation'])
    roomBuilder.update_infiltration_rel(params['infiltration'])
    roomBuilder.updateRoomSize(params['width'], params['length'], params['height'])
    params['_floorArea'] = roomBuilder.calculateFloorArea()
    params['_volume'] = round(params['width'] * params['length'] * params['height'], 3)
    params['_exteriorSurfaceArea'] = roomBuilder.calculateExteriorWallArea()
    params['_winToFloorRatio'] = roomBuilder.calculateWindowToFloorRatio()
    return params

def buildRooms(roomBuilder, samples, winToFloor_bounds, rooms=10):
    """
    Uses randomized sets of room parameters based on samples from a previous LHS
    to build the respective idf models via a passed roomBuilder instance.
    Samples are ignored if the required lighting condition for a valid room is not met 
    (window-to-floor area ratio between the specified bounds).
    The function keeps generating new rooms until a passed number of valid rooms is reached.
    :param roomBuilder: modelling instance to generate a room model
    :param sampels: samples from a previous lhs
    :param winToFloor_bounds: acceptable range of ratio between window area and floor area [min, max]
    :param rooms: number of room models to generate
    :return: 
    """
    validParameterSets = []
    for i in range(len(samples)):
        if len(validParameterSets) == rooms:
            break
        params = {}
        params['width']         = samples[i, 0]
        params['length']        = samples[i, 1]
        params['height']        = samples[i, 2]
        params['infiltration']  = samples[i, 3]
        params['outdoor_co2']   = samples[i, 4]
        params['orientation']   = samples[i, 5]
        params['maxOccupants']  = samples[i, 6]
        params = buildRoomModel(roomBuilder, params)
        if ((params['_winToFloorRatio'] >= winToFloor_bounds[0]) & (params['_winToFloorRatio'] <= winToFloor_bounds[1])):
            # only add room if it matches the required lightning condition
            validParameterSets.append(params)
    return validParameterSets
```

```python
# build 50000 rooms that match the lightning constraints
rb = RoomBuilder(idfFile, iddFile, idf_output_path)
rooms = buildRooms(rb, samples, winToFloor_bounds, rooms=50000)
len(rooms)
```

### Populate Rooms

```python
df_occupancy = pd.read_csv(occupancyFile) # load occupancy data from file

def read_occupancy_day_profiles(maxOccupants, field='Occupants', use_fractions=True):
    """
    Reads occupancy data from the dataframe.
    Returns the data as a 2D-array representing days with 1440 values each (one value per minute)
    :param maxOccupants: maximum number of occupants to filter for
    :param field: column name in the csv
    :param use_fractions: boolean, whether the absolute values shall be used (False) 
                                   or a fraction of the maximum value (True) 
    """
    df = df_occupancy
    df = df[df.maxOccupants == maxOccupants]
    values = df[field]
    if use_fractions: # use fractions instead of absolute numbers
        values = values / maxOccupants 
    values = np.array(values).reshape(-1, 1440) # array of days with 1440 values each
    return values

def pickOccupantData(occupancy_data, ventilation_data, days=7):
    """
    Picks a random sequence of days from the occupancy and ventilation data 
    according to the passed length in days (by default 7 days).
    Returns a tuple of the sequence's first day index and the sequence.
    """
    i = random.randint(0, len(occupancy_data)-days)
    return i, occupancy_data[i:i+days], ventilation_data[i:i+days]         
            
def populateRoom(params, simulation_days):
    """
    Selects the maximum number of occupants for the room (meeting the restriction of m² per occupant).
    According to this number and the passed number of simulation days, this function 
    picks and returns a random sequence of occupancy and ventilation data 
    (together with an updated parameter dictionary).
    :param params: dictionary of room parameters
    :param simulaton_days: number of days of occupancy/ventilation data to read
    """
    occupancy_data = read_occupancy_day_profiles(params['maxOccupants'])
    ventilation_data = read_occupancy_day_profiles(params['maxOccupants'], 
                                                   'WindowOpeningFraction', False)
    params['firstDayOfUsedOccupancySequence'], occupancy_sequence, ventilation_sequence = \
            pickOccupantData(occupancy_data, ventilation_data, simulation_days)  
    return params, occupancy_sequence, ventilation_sequence
        
params = df_rooms.iloc[0, :].to_dict()
params, occupancy_sequence, ventilation_sequence = populateRoom(params, simulation_days)
params
```

### Simulation Functions

```python
def randomize_runperiod(days=10, year=2023):
    '''
    Picks a random simulation runperiod. 
    :param days: length of the runperiod in days
    :param year: year within which the period is picked
    '''
    d = random.randint(1, 365-days+1)
    start_date = (datetime(year, 1, 1) + timedelta(days=d-1))
    runperiod = (start_date.strftime("%Y-%m-%d"), (start_date + timedelta(days=days-1)).strftime("%Y-%m-%d"))
    return runperiod

def simulate(days=10):
    '''Conducts an EnergyPlus simulaton over a passed number of days'''
    
    # create idf object
    idf = IDF(idf_output_path, epwFile)
    print(idf.idfname.split('.')[0])

    # set time step to 1 min
    idf.idfobjects['TIMESTEP'][0].Number_of_Timesteps_per_Hour = 60

    # set options to same behavior as EP-Launch
    ep_options = {
            'output_prefix': idf.idfname.split('.')[-2].split('/')[-1],
            'output_suffix': 'C',
            'output_directory': os.path.dirname(idf.idfname),
            'expandobjects':True,
            'readvars':True
    }

    # set outdoor co2 schedule and request output of CO2 concentration
    add_constant_outdoor_co2_schedule(idf, params['outdoor_co2'])
    request_output(idf, "Zone Air CO2 Concentration")
    request_output(idf, "Site Outdoor Air Barometric Pressure")

    # change runperiod
    runperiod = randomize_runperiod(days)
    print("Simulated period:", runperiod)
    set_runperiod(idf, runperiod[0], runperiod[1])

    # set occupancy
    set_occupancy(idf, zoneName, runperiod, occupancy_sequence,
                   params['maxOccupants'], generationrate_co2, activity_level, True)

    # set ventilation
    set_ventilation(idf, zoneName, runperiod, ventilation_sequence)

    # run simulation
    idf.run(**ep_options)
    idf.saveas("ep_data/room_for_execution.idf")
```

```python
def read_results(idf_output_path):
    '''Reads the results from a conducted simulation and returns them in a pandas DataFrame'''
    df = pd.read_csv('ep_data/Output/eplusout.csv')
    column_names = [c.split(':')[1].split(' [')[0] for c in [c for c in df.columns if ':' in c]]
    column_names.insert(0, "Datetime")
    df.columns = column_names
    df['Occupancy'] = np.array(occupancy_sequence).reshape(-1)[:len(df)]
    df['Ventilation'] = np.array(ventilation_sequence).reshape(-1)[:len(df)]
    # check if occupancy was correctly considered during simulation
    if len(df[(df['Zone People Total Heating Energy']>0) != (df['Occupancy']>0)]) > 0:
        print("There was an error. Occupancy was not correctly simulated!")
        return None
    df = df[['Datetime', 'Zone Air CO2 Concentration', 'Zone Mean Air Temperature', 'Zone Air Relative Humidity', 'Occupancy', 'Ventilation']]
    return df
```

### Simulate 50000 Rooms

```python
def count_datapoints():
    try:
        with h5py.File(h5_path, 'r') as f:
            return len(f['data/table'])
    except:
        return 0
    
def read_last_simID():
    return pd.read_hdf(h5_path, key='metadata')['simID'].max()
```

```python
# create new hdf5 file if not existing yet
if not os.path.exists(h5_path):
    f = h5py.File(h5_path, 'a')
    f.close()
    firstSimID = 0
else:
    firstSimID = read_last_simID() + 1

# run 50000 simulations
for simID in range(firstSimID, 50000):
    
    print("Iteration", simID)
    
    # read room parameters
    params = df_rooms.iloc[simID, :].to_dict()
    
    # build room model and run simulation
    rb = RoomBuilder(idfFile, iddFile, idf_output_path)
    params = buildRoomModel(rb, params) # build the room model and save a temporary idf file for simulation
    params, occupancy_sequence, ventilation_sequence = populateRoom(params, simulation_days)
    print(params)
    simulate(days=simulation_days) # does prepare idf but not run the simulation yet
    
    # read simulation results 
    try:
        df = read_results(idf_output_path)
        print("Simulation finished with {} simulated days".format(int(len(df)/1440)))
        print("CO2 min-median-max:", round(df['Zone Air CO2 Concentration'].min(),1), round(df['Zone Air CO2 Concentration'].median(), 1), round(df['Zone Air CO2 Concentration'].max(),1))
        print("Temperature min-median-max:", round(df['Zone Mean Air Temperature'].min(),1), round(df['Zone Mean Air Temperature'].median(),1), round(df['Zone Mean Air Temperature'].max(),1))
    except:
        simID -= 1
        continue
        
    # save
    df['simID'] = simID
    params['simID'] = simID
    df['BinaryOccupancy'] = df['Occupancy'].apply(lambda x: 1 if x > 0 else 0)
    with pd.HDFStore(h5_path, mode='a') as store:
        store.put('data', df, format='t', append=True, index=False)
        store.put('metadata', pd.DataFrame([params]), format='t', append=True, index=False)
    print("___")
    print("Stored simulated data to file " + h5_path)
    dp = count_datapoints()
    print("File now contains {} datapoints from {} simulations".format(dp, int(dp / (simulation_days * 1440))))
    print("___")
    print("")
    
print("Done")
```
